```python
# Select numerical efficiency indicator columns
numerical_efficiency_cols = [
    'Standardized ED visits ratio (Facility)',
    'Standardized ED visits within 30 days of hospital discharge ratio (Facility)',
    'Standardized First Kidney Transplant Waitlist Ratio',
    'Fistula Rate (Facility)',
    'Percent of Adult HD patients with Kt/V >= 1.2',
    'Percentage of Adult PD PTS with Kt/V >= 1.7',
    'Percentage of Adult patients with hypercalcemia (serum calcium greater than 10.2 mg/dL)',
    'Percentage of Adult patients with long term catheter in use',
    'Percentage of Adult patients with serum phosphorus between 3.5-4.5 mg/dL',
    'Percentage of Adult patients with serum phosphorus between 4.6-5.5 mg/dL',
    'Percentage of Adult patients with serum phosphorus between 5.6-7.0 mg/dL',
    'Percentage of Adult patients with serum phosphorus greater than 7.0 mg/dL',
    'Percentage of Medicare patients with Hgb<10 g/dL',
    'Percentage of Medicare patients with Hgb>12 g/dL',
    'Percentage of Pediatric HD patients with Kt/V >= 1.2',
    'Percentage of Prevalent Patients Waitlisted',
    'Percentage of pediatric HD patients with nPCR',
    'Percentage of pediatric PD patients with Kt/V>=1.8',
    'Readmission Rate (Facility)',
    'Mortality Rate (Facility)'
]

survey_numerical_efficiency = survey[numerical_efficiency_cols].copy()
display(survey_numerical_efficiency.head())
```

```python
# Select categorical efficiency indicator columns
categorical_efficiency_cols = [
    'Patient Hospital Readmission Category',
    'Patient Infection category text',
    'Patient Transfusion category text',
    'Patient hospitalization category text'
]

survey_categorical_efficiency = survey[categorical_efficiency_cols].copy()
display(survey_categorical_efficiency.head())
```

```python
# Fill missing values in numerical efficiency columns with the mean
survey_numerical_efficiency = survey_numerical_efficiency.fillna(survey_numerical_efficiency.mean())
display(survey_numerical_efficiency.isnull().sum())
```

```python
# Fill missing values in categorical efficiency columns with the mode
for col in survey_categorical_efficiency.columns:
    survey_categorical_efficiency[col] = survey_categorical_efficiency[col].fillna(survey_categorical_efficiency[col].mode()[0])

display(survey_categorical_efficiency.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler

# Scale numerical efficiency columns
scaler = StandardScaler()
survey_numerical_scaled = scaler.fit_transform(survey_numerical_efficiency)
survey_numerical_scaled = pd.DataFrame(survey_numerical_scaled, columns=numerical_efficiency_cols)
survey_numerical_scaled['State'] = survey['State'] # Add State column for grouping
display(survey_numerical_scaled.head())
```

```python
# Map categorical values to numerical scores
category_mapping = {
    'Better than Expected': 1,
    'As Expected': 0,
    'Worse than Expected': -1,
    'Not Available': 0 # Assuming 'Not Available' is neutral
}

survey_categorical_scaled = survey_categorical_efficiency.replace(category_mapping)
survey_categorical_scaled['State'] = survey['State'] # Add State column for grouping

display(survey_categorical_scaled.head())
```

```python
# Group by state and calculate the mean of the scaled numerical efficiency scores
preliminary_efficiency_score = survey_numerical_scaled.groupby('State').mean()

# Group by state and calculate the mean of the scaled categorical efficiency scores
categorical_efficiency_score = survey_categorical_scaled.groupby('State').mean()

# Combine the scores
combined_scores = pd.merge(preliminary_efficiency_score, categorical_efficiency_score, on='State')

# Calculate a simple composite score by averaging all the scaled scores
combined_scores['Composite Efficiency Score'] = combined_scores.mean(axis=1)

composite_efficiency_df = combined_scores[['Composite Efficiency Score']]

display(composite_efficiency_df.head())
```

```python
# Sort states by composite efficiency score in descending order
most_efficient_states = composite_efficiency_df.sort_values(by='Composite Efficiency Score', ascending=False)

# Display the most efficient states (e.g., top 10)
display(most_efficient_states.head(10))
```

## 2. My Approach to Defining "Efficiency" for Dialysis Centres

**Definition:** Efficiency for dialysis centers in this context is defined as a composite measure reflecting positive patient outcomes and lower risk of adverse events, achieved with potentially fewer resources or better resource utilization (though resource utilization is indirectly inferred from outcomes like readmission rates). A higher efficiency score indicates better performance.

The score is calculated based on the aggregate of the following metrics:

```python
# Check for missing values
print("Missing values before handling:")
display(survey.isnull().sum().head())

# Check for duplicate rows
print("\nNumber of duplicate rows:", survey.duplicated().sum())

# Handle missing values (example: fill numerical NaNs with mean, categorical NaNs with mode)
for col in survey.columns:
    if survey[col].isnull().any():
        if survey[col].dtype in ['float64', 'int64']:
            survey[col] = survey[col].fillna(survey[col].mean())
        else:
            survey[col] = survey[col].fillna(survey[col].mode()[0])

# Drop duplicate rows
survey = survey.drop_duplicates()

# Verify missing values and duplicates after handling
print("\nMissing values after handling:")
display(survey.isnull().sum().head())
print("\nNumber of duplicate rows after handling:", survey.duplicated().sum())
```
